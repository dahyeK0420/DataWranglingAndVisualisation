```python
df = pd.ExcelFile('jp_FoodG_male_2012.xlsx').parse(0)
```

```python
def droppingColIndex(df):
    lst = []
    for i in range(3,df.shape[1],3): 
        lst.append(i)
        lst.append(i+1)
    return lst
```

```python
age = list(df.loc[5,:][df.loc[5,:].notna()])[2:]
```

```python
df.dropna(axis = 0, inplace = True)
```

```python
df.drop(df.columns[droppingColIndex(df)], axis = 1, inplace = True)
```

```python
df.drop(['Unnamed: 1'], axis = 1, inplace = True)
df.columns = ['Elements','All']+age
```

```python
def parsingDemo(df): 
    age = list(df.loc[5,:][df.loc[5,:].notna()])[2:]
    df.dropna(axis = 0, inplace = True)
    
    df.drop(df.columns[droppingColIndex(df)], axis = 1, inplace = True)
    df.drop(['Unnamed: 1'], axis = 1, inplace = True)
    df.columns = ['Elements','All']+age
    return df
```

```python
automatable = [i for i in os.listdir() if re.findall('2012|2013|2014|2015|2016',i)]
```

```python
for i in automatable: 
    if re.findall('_male_',i):
        print(i)
        df = pd.ExcelFile(i).parse(0)
        age = list(df.loc[5,:][df.loc[5,:].notna()])[2:]
        df.dropna(axis = 0, inplace = True)

        df.drop(df.columns[droppingColIndex(df)], axis = 1, inplace = True)
        df.drop(['Unnamed: 1'], axis = 1, inplace = True)
        df.columns = ['Elements','All']+age
```

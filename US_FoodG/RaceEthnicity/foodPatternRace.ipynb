```python
import os 
import re 
import pandas as pd 
import tabula
import numpy as np
```

```python
race = sorted([i for i in os.listdir() if re.findall('race.pdf',i)])
```

```python
race
```

```python
fruit = ['total fruit', 'citrus, melons, berries', 'other fruit']
dfs = tabula.read_pdf(race[0], pages = 1)
```

```python
race1 = ['Non-Hispanic White']*5+ ['Non-Hispanic Black']*5+['Mexican American']*5+['All Hispanic']*5
race2 = ['Non-Hispanic White']*5+ ['Non-Hispanic Black']*5+['Non-Hispanic Asian']*5+['All Hispanic']*5
df = dfs[0].dropna(axis = 1, thresh = 10)
```

```python
def reorganising_tempF(df): 
    
    df.dropna(inplace = True, axis = 1, thresh = 10)
    df['temp'] = ['']*df.shape[0]
    for i in range(df.shape[1]): 
        df['temp']+=df[df.columns[i]]
        useless = [i for i in range(0,df.shape[1]-1)]
    df.drop(df.columns[useless], axis = 1, inplace = True)

    df['temp']=df['temp'].apply(lambda x: re.sub('^\D.*','',x))
    df.temp = df.temp.apply(lambda x: x.replace('(', ' '))
    df.temp = df.temp.apply(lambda x: x.replace(')', ' '))
    df.temp = df.temp.apply(lambda x: x.replace('*', ' '))
    df = df[df.temp!='']
    
    return df 
```

```python
def allocateNumbersF(df, yr):
    
    for i in range(0, len(fruit)): 
        df[fruit[i]] = df['temp'].apply(lambda x: x.split('... ')[1].split()[i*2+1])
        
    df['age'] = df[df.columns[0]].apply(lambda x: x.split('... ')[0].rstrip('...'))
    df['year'] = [yr]*df.shape[0]
    if int(yr.split('-')[0])<2011: 
        lst = race1
    else: lst = race2
    df['race'] = lst
    df.drop(['temp'], axis = 1, inplace = True)
    
    return df
```

```python
raceFruits =[]
for i in range(0, len(race)): 
    year = race[i].split('-')[0]+'-'+str(int(race[i].split('-')[0])+1)
    dfs = tabula.read_pdf(race[i], pages = 1)
    tempdf = reorganising_tempF(dfs[0])
    tempdf = allocateNumbersF(tempdf,year)
    raceFruits.append(tempdf)
```
